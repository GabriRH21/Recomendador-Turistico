## Carga de datos

En primer lugar extraemos los datos del fichero (categoria y respectivo logaritmo). El pop se hace para evitar que queden filas vaicas al final del fichero. Comenzamos un bucle que recorrera el fichero de 8 en 8 filas, ya que cada activida consume 8 filas. El if es para entrar solo en las lineas que tiene los logaritmos. Con otro bucle recorremos las lineas para quedarnos unicamente con los numeros

```python

with open('classificationInfo.txt', 'r') as file:
    # Lee las líneas del archivo
    lines = file.readlines()
    with open("percentages.txt", 'w') as n:
      iterator = 1
      while lines[-1] == "\n":
        lines.pop()
      #print(lines[-1])
      for i in range(0, len(lines), 8):
        Logarithms = {}
        # Procesa las 8 líneas
        for j in range(8):
            if i + j < len(lines):
              if j > 0 and j < 6:
                #print(lines[i + j].strip())  # Imprime la línea sin el salto de línea
                words = lines[i + j].split()
                category = words[0][:-1]          # quita los :
                #print(category)
                for word in words:
                    try:
                        number = float(word)
                        #print(number)
                        Logarithms[category] = number

                    except ValueError:
                        pass
        result = discard_and_order(Logarithms)
        print(result)
        n.write("____Porcentajes de la actividad " + str(iterator) + "____: \n" )
        for catego, numb in  result.items():
          n.write("Categoria: " + catego + ". Porcentaje: " + str(numb) + "\n")
        n.write("\n-------\n\n")
        iterator += 1

        print("-------")  # Separador entre bloques de 8 líneas
```
